```python
import numpy as np
import random

# Defina a matriz de distância entre as cidades
# Substitua esta matriz com os dados do seu arquivo TSP
distance_matrix = np.array([[1, 0, 13],
[2, 0, 26],
[3, 0, 27],
[4, 0, 39]
[5, 2, 0]
[6, 5, 13]
[7, 5, 19]
[8, 5, 25]
[9, 5, 31]
[10, 5, 37]
[11, 5, 43]
[12, 5, 8]
[13, 8, 0]
[14, 9, 10]
[15, 10, 10]
[16, 11, 10]
[17, 12, 10]
[18, 12, 5]
[19, 15, 13]
[20, 15, 19]
[21, 15, 25]
[22, 15, 31]
[23, 15, 37]
[24, 15, 43]
[25, 15, 8]
[26, 18, 11]
[27, 18, 13]
[28, 18, 15]
[29, 18, 17]
[30, 18, 19]
[31, 18, 21]
[32, 18, 23]
[33, 18, 25]
[34, 18, 27]
[35, 18, 29]
[36, 18, 31]
[37, 18, 33]
[38, 18, 35]
[39, 18, 37]
[40, 18, 39]
[41, 18, 41]
[42, 18, 42]
[43, 18, 44]
[44, 18, 45]
[45, 25, 11]
[46, 25, 15]
[47, 25, 22]
[48, 25, 23]
[49, 25, 24]
[50, 25, 26]
[51, 25, 28]
[52, 25, 29]
[53, 25, 9]
[54, 28, 16]
[55, 28, 20]
[56, 28, 28]
[57, 28, 30]
[58, 28, 34]
[59, 28, 40]
[60, 28, 43]
[61, 28, 47]
[62, 32, 26]
[63, 32, 31]
[64, 33, 15]
[65, 33, 26]
[66, 33, 29]
[67, 33, 31]
[68, 34, 15]
[69, 34, 26]
[70, 34, 29]
[71, 34, 31]
[72, 34, 38]
[73, 34, 41]
[74, 34, 5]
[75, 35, 17]
[76, 35, 31]
[77, 38, 16]
[78, 38, 20]
[79, 38, 30]
[80, 38, 34]
[81, 40, 22]
[82, 41, 23]
[83, 41, 32]
[84, 41, 34]
[85, 41, 35]
[86, 41, 36]
[87, 48, 22]
[88, 48, 27]
[89, 48, 6]
[90, 51, 45]
[91, 51, 47]
[92, 56, 25]
[93, 57, 12]
[94, 57, 25]
[95, 57, 44]
[96, 61, 45]
[97, 61, 47]
[98, 63, 6]
[99, 64, 22]
[100, 71, 11]
[101, 71, 13]
[102, 71, 16]
[103, 71, 45]
[104, 71, 47]
[105, 74, 12]
[106, 74, 16]
[107, 74, 20]
[108, 74, 24]
[109, 74, 29]
[110, 74, 35]
[111, 74, 39]
[112, 74, 6]
[113, 77, 21]
[114, 78, 10]
[115, 78, 32]
[116, 78, 35]
[117, 78, 39]
[118, 79, 10]
[119, 79, 33]
[120, 79, 37]
[121, 80, 10]
[122, 80, 41]
[123, 80, 5]
[124, 81, 17]
[125, 84, 20]
[126, 84, 24]
[127, 84, 29]
[128, 84, 34]
[129, 84, 38]
[130, 84, 6]
[131, 107, 27])

# Parâmetros do algoritmo genético
population_size = 50
generations = 1000
mutation_rate = 0.01

# Função de fitness: calcular a distância total da rota
def calculate_total_distance(route):
    total_distance = 0
    for i in range(len(route) - 1):
        total_distance += distance_matrix[route[i]][route[i+1]]
    return total_distance + distance_matrix[route[-1]][route[0]]  # Volta à cidade inicial

# Inicialização da população
population = [list(range(len(distance_matrix))) for _ in range(population_size)]

# Algoritmo Genético
for generation in range(generations):
    # Avaliação da população
    fitness_scores = [1/calculate_total_distance(route) for route in population]
    
    # Seleção dos pais (roleta)
    parents = np.random.choice(population, size=population_size//2, p=fitness_scores/np.sum(fitness_scores), replace=True)
    
    # Cruzamento (ordem)
    children = []
    for i in range(0, len(parents), 2):
        parent1, parent2 = parents[i], parents[i+1]
        crossover_point = random.randint(0, len(parent1) - 1)
        child1 = parent1[:crossover_point] + [city for city in parent2 if city not in parent1[:crossover_point]]
        child2 = parent2[:crossover_point] + [city for city in parent1 if city not in parent2[:crossover_point]]
        children.extend([child1, child2])
    
    # Mutação
    for i in range(len(children)):
        if random.random() < mutation_rate:
            mutation_point1, mutation_point2 = random.sample(range(len(children[i])), 2)
            children[i][mutation_point1], children[i][mutation_point2] = children[i][mutation_point2], children[i][mutation_point1]
    
    # Substituir a população pelos filhos
    population = children

# Encontrar a melhor solução
best_route = min(population, key=calculate_total_distance)
best_distance = calculate_total_distance(best_route)

print("Melhor rota encontrada:", best_route)
print("Distância total da melhor rota:", best_distance)
```
